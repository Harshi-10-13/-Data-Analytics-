If condition

```python
age = int(input("Enter your age: "))

if age >= 18:
    print("Eligible to vote")
else:
    print("Not eligible")
```

For Loop

```python
for i in range(1, 50):
    print(i)
```

```python
for i in range(1, 21):
    if i % 2 == 0:
        print(i)
```

Nested If Loop

```python
num = int(input("Enter number: "))

if num > 0:
    if num % 2 == 0:
        print("Positive Even")
    else:
        print("Positive Odd")
else:
    print("Negative number")
```

While Loop

```python
i = 1

while i <= 5:
    print(i)
    i  += 1
```

Square Using If Condition

```python
num = int(input("Enter number: "))

if num > 0:

    print("Square is:", num ** 2)
else:
    print("Enter positive number")
```

```python
def greet():
    print("My name is Harshita")

greet()
```

Upper and Lower Case

```python
name = "Harshita"

print(name.upper())
print(name.lower())
```

String Slicing

```python
word = "Python"

print(word[0:3])
print(word[-1])
print(word[::-1])
```

Reverse Slicing

```python
text = "Harshita"

print(text[::-4])
```

Start, Stop, Step

```python
name = "Programming"

print(name[0:10:2])
print(name[1:10:4])
print(name[::-2])
```

Reverse with Step

```python
data = "Python"

print(data[::-1])
print(data[::-3])
```

List Comprehension

```python
numbers = [x for x in range(1, 11)]

print(numbers)
```

Comprehension

Square Using Comprehension

```python
square = [x**3 for x in range(1, 10)]

print(square)
```

```python
even = [x for x in range(1, 21) if x % 2 == 0]

print(even)
```

```python
odd = [x for x in range(1, 21) if x % 2 != 0]

print(odd)
```

```python
student = {"name": "Harshita"}

student.update({"age": 20})

print(student)
```
