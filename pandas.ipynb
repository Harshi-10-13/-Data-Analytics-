Null Value

```python
import pandas as pd

data = {
    "Name": ["A", "B", None],
    "Marks": [90, None, 80]
}

df = pd.DataFrame(data)

print(df.isnull())
```

Average mean

```python
import pandas as pd

data = {
    "Marks": [80, 90, 70, 60]
}

df = pd.DataFrame(data)

print(df["Marks"].mean())
```

fillna

```python
import pandas as pd

data = {
    "Marks": [90, None, 80, None]
}

df = pd.DataFrame(data)

df["Marks"] = df["Marks"].fillna(0)

print(df)
```

mode

```python
import pandas as pd

data = {
    "Marks": [90, 80, 90, 70, 90]
}

df = pd.DataFrame(data)

print(df["Marks"].mode())
```

inplace

```python
import pandas as pd

data = {
    "Name": ["A", "B", "C"],
    "Marks": [80, 90, 70]
}

df = pd.DataFrame(data)

df.drop("Marks", axis=1, inplace=True)

print(df)
```

dropna

```python
import pandas as pd

data = {
    "Name": ["A", None, "C"],
    "Marks": [80, 90, None]
}

df = pd.DataFrame(data)

new_df = df.dropna()

print(new_df)
```
